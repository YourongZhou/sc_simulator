```python
import numpy as np
import pandas as pd
import scanpy as sc
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
adata = sc.read_h5ad('sim_data/adata/mule_sim.h5ad')
```

```python
proportion = np.count_nonzero(adata.X, axis=0) / adata.shape[0]
median_per_gene = np.median(adata.X, axis=0)
mean_per_gene = np.mean(adata.X, axis=0)
```

```python
plt.scatter(proportion, np.log(median_per_gene+1))
```

```python
plt.scatter(proportion, np.log(mean_per_gene+1))
```
